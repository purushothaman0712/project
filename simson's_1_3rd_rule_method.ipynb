# **Calculate flow of quantity in fluid dynamics (using the simpsons 1/3rd rule)**

**Fluid Dynamics**

Fluid flow through pipes or channels often involves non-linear relationships between various parameters. Simpson's Rule proves useful in such scenarios. It can be used to estimate the average fluid velocity across a pipe's cross-section or the total volume of fluid flowing over time. This plays a crucial role in designing efficient pipelines or analyzing fluid behavior in engines, where understanding flow rates and volumes is essential.It is used to estimate the average fluid velocity across a pipe's cross-section, flow rates, or the total volume of fluid flowing over time in complex systems like engines or pipelines where flow parameters are non-linear.

```python
import numpy as pu
import matplotlib.pyplot as blk
def f(x):
  return x**2                                 # velocity function
lower=0                                       # lower time limits
upper=4                                       # upper time limits
print("n must be even for Simpson's Rule")
subintervel=4                                 # number of equal intervals (must be even)
stepsize=(upper-lower)/subintervel            # width of each interval
sum_val=f(upper)+f(lower)                     # initialize the summation
# loop through intermediate points
for i in range(subintervel):
  k=lower+stepsize*i
  # apply simpsons's coefficients
  if(i%2==0):                                 # even index multiply by 2
    sum_val=sum_val+2*f(k)
  else:                                       #odd index multiply 4
    sum_val=sum_val+4*f(k)
integration=stepsize/3*sum_val                # final integration formula
# display the result
print("flow of quantity:",integration)        # this prints flow(area under velocity-time curve)


```

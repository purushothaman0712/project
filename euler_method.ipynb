Kinematics and Dynamics: Simulating object motion where acceleration is not constant, such as falling objects with air resistance (viscous drag), projectile motion, and rocket propulsion

# simulate vertical rocket motion using Euler’s method

Rocket motion involves variable mass and non-constant acceleration. Euler’s method numerically integrates the coupled equations of motion for velocity, position, and mass, enabling simulation of realistic rocket flight under thrust and gravity.

```python
import matplotlib.pyplot as plt

# Physical constants
g = 9.8            # gravity (m/s^2)
ve = 2000          # exhaust velocity (m/s)
burn_rate = 5      # fuel burn rate (kg/s)

# Masses
m0 = 1000         # initial mass (rocket + fuel) in kg
mf = 700          # final mass after fuel burn (kg)

# Time parameters
dt = 0.2          # small time step (Euler method requirement)
t_max = 60        # total simulation time

# Initial conditions
t = 0.0           # mass
v = 100           # initial velocity (m/s)
x = 10            # initial height (m)
m = m0            # mass

# Lists to store data
time = []
height = []
velocity = []
mass = []

# Euler loop
# until time reaches t_max
# until fuel is exhausted (m > mf)
while t <= t_max and m > mf:
    # Store current values
    time.append(t)
    height.append(x)
    velocity.append(v)
    mass.append(m)

    # Mass change
    dm_dt = -burn_rate     # Negative sign → mass is decreasing (fuel is burning)

    # Acceleration from rocket equation
    a = (ve * (-dm_dt) / m) - g

    # Euler updates
    v = v + a * dt       # kinematics
    x = x + v * dt       #  Euler update for position
    m = m + dm_dt * dt   # Mass decreases every step
    t = t + dt           # Move to next time step

# Plot results
plt.figure(figsize=(10,5))
plt.plot(time, height, label="Height (m)")
plt.plot(time, velocity, label="Velocity (m/s)")
plt.plot(time, mass, label="Mass (kg)")
plt.xlabel("Time (s)")
plt.ylabel("Value")
plt.title("Rocket Motion using Euler Method")
plt.legend()
plt.grid(True)
plt.show()
```

#Falling Body with Air Resistance (Euler Method)

A falling body with air resistance experiences velocity-dependent force, leading to non-constant acceleration. Euler’s method numerically integrates the equations of motion by updating velocity and position in small time steps, allowing realistic simulation of the fall and prediction of terminal velocity

```python
import matplotlib.pyplot as plt

# Physical parameters
m = 1.0       # mass (kg)
k = 0.8       # drag constant
g = 9.8       # gravity (m/s^2)

# Time parameters
dt = 0.01     # time step
t_max = 10    # total time

# Initial conditions
t = 0.0       # time
v = 0.0       # initial velocity
y = 0.0       # initial height

# Lists to store values
time = []
velocity = []
position = []

# Euler loop
# Loop runs until the total time is reached
while t <= t_max:
    time.append(t)
    velocity.append(v)
    position.append(y)

    # Acceleration (from dynamics)
    a = g - (k/m) * v

    # Euler updates (from kinematics)
    v = v + a * dt    # Euler formula for velocity
    y = y + v * dt    # Euler formula for position
    t = t + dt        # Moves to the next time step

# Plot graphs
plt.plot(time, velocity, label="Velocity (m/s)")
plt.plot(time, position, label="Position (m)")
plt.xlabel("Time (s)")
plt.ylabel("Value")
plt.title("Falling Body with Air Resistance (Euler Method)")
plt.legend()
plt.grid(True)
plt.show()
```

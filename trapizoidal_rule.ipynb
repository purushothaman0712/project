#Calculate area under the curve an experimental graph (using the trapezoidal rule)

Trapezium Rule (Trapezoidal Rule): This is a
more formal numerical method that divides the area under the curve into a series of vertical strips (trapezoids) and sums their areas. This provides a better approximation than counting squares.

```python
#experimental data (displacement vs force) to find the energy
import numpy as pu
import matplotlib.pyplot as blk
n=int(input("enter the number of elements:"))
x=[]
y=[]
for i in range(n):
  x.append(int(input(f"enter the x value x{i+1}:")))
  y.append(int(input(f"enter the y value y{i+1}:")))
energy=pu.trapz(y,x)
print("enter (area under curve)=",energy,"joules")
blk.plot(x,y,marker="o")
blk.xlabel("Displacement (meters)")
blk.ylabel("Force (newton)")
blk.title(f"Energy={energy}joules")
blk.fill_between(x,y, alpha=0.3)
blk.show()
```
